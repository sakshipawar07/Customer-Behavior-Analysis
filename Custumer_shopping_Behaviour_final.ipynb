```python
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x :x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df = df.rename(columns = {"purchase_amount_(usd)":"purchase_amount"})
```

```python
def age_group(age):
    if age <= 30 :
        return "Young adult"
    elif age <=50:
        return "middle aged"
    else:
        return "senoir"
```

```python
# create new column age group

# df["age_group"] = df["age"].apply(age_group)
  
# or using qcut    # divides data into eual bins

df["age_group"] = pd.qcut(df["age"],q=3,labels=["young adult","middle age","senior"])
```

```python
# create col purchase_freq_days

freq_mapping = {
    "Annually": 365,
    "Bi-Weekly":14,
    "Every 3 Months":90,
    "Quarterly":90,
    "Monthly":30,
    "Weekly":7,
    "Fortnightly":14
}

df["purchase_freq_days"] = df["frequency_of_purchases"].map(freq_mapping)
```
